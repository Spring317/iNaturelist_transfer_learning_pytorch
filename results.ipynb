```python
import matplotlib.pyplot as plt
import numpy as np
Conv_acc =[0.953, 0.939,0.947]
Conv_throughput = [8.36, 8.4, 8.53]
D1_acc  =[0.898, 0.8700, 0.888]
D1_throughput = [21.63, 11.41, 10.09]
# D1_fn_rate = [0.027, 0.027, 0.027]

D2_acc  =[ 0.8960, 0.8680, 0.881]
D2_throughput = [21.64, 17.61, 12.58]
# D2_fn_rate = [0.062, 0.062, 0.062]

D3_acc  =[ 0.8920, 0.8640, 0.876]
D3_throughput = [22.03, 17.36, 15.18]

# Setup for grouped bar chart
x = np.arange(3)  # 4 experiments
width = 0.2  # width of bars

# First Chart - Accuracy
plt.figure(figsize=(10, 6))
plt.bar(x - width, Conv_acc, width, label='Conv', color='#d62728', alpha=0.8)
plt.bar(x, D1_acc, width, label='1 pipe', color='#1f77b4', alpha=0.8)
plt.bar(x + width, D2_acc, width, label='2 pipe', color='#ff7f0e', alpha=0.8)
plt.bar(x + 2 * width, D3_acc, width, label='3 pipe', color='#2ca02c', alpha=0.8)
# Add value labels on bars
for i, (d1, d2, d3, d4) in enumerate(zip(Conv_acc, D1_acc, D2_acc, D3_acc)):
    plt.text(i - width, d1 + 0.01, f'{d1:.3f}', ha='center', va='bottom', fontsize=9)
    plt.text(i, d2 + 0.01, f'{d2:.3f}', ha='center', va='bottom', fontsize=9)
    plt.text(i + width, d3 + 0.01, f'{d3:.3f}', ha='center', va='bottom', fontsize=9)
    plt.text(i + 2 * width, d4 + 0.01, f'{d4:.3f}', ha='center', va='bottom', fontsize=9)
    

plt.title('Accuracy Comparison', fontweight='bold', fontsize=14)
plt.ylabel('Accuracy')
plt.xlabel('Experiment')
plt.xticks(x, ['DI1', 'DI2', 'DI3'])
plt.legend()
plt.grid(True, alpha=0.3)
# plt.ylim(0.85, 0.90)
plt.tight_layout()
plt.show()

# Second Chart - Throughput
plt.figure(figsize=(10, 6))
plt.bar(x - width, Conv_throughput, width, label='Conv', color='#d62728', alpha=0.8)
plt.bar(x, D1_throughput, width, label='D1', color='#1f77b4', alpha=0.8)
plt.bar(x + width, D2_throughput, width, label='2 pipe', color='#ff7f0e', alpha=0.8)
plt.bar(x + 2 * width, D3_throughput, width, label='3 pipe', color='#2ca02c', alpha=0.8)

for i, (d1, d2, d3, d4) in enumerate(zip(Conv_throughput, D1_throughput, D2_throughput, D3_throughput)):
    plt.text(i - width, d1 + 0.01, f'{d1:.3f}', ha='center', va='bottom', fontsize=9)
    plt.text(i, d2 + 0.01, f'{d2:.3f}', ha='center', va='bottom', fontsize=9)
    plt.text(i + width, d3 + 0.01, f'{d3:.3f}', ha='center', va='bottom', fontsize=9)
    plt.text(i + 2 * width, d4 + 0.01, f'{d4:.3f}', ha='center', va='bottom', fontsize=9)


plt.title('Throughput Comparison', fontweight='bold', fontsize=14)
plt.ylabel('Throughput (fps)')
plt.xlabel('Experiment')
plt.xticks(x, ['DI1', 'DI2', 'DI3'])
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()


# Print summary statistics
print("Summary Statistics:")
print("=" * 50)
print(f"D1 - Accuracy: {np.mean(D1_acc):.3f}  {np.std(D1_acc):.3f}, Throughput: {np.mean(D1_throughput):.1f}  {np.std(D1_throughput):.1f}")
print(f"D2 - Accuracy: {np.mean(D2_acc):.3f}  {np.std(D2_acc):.3f}, Throughput: {np.mean(D2_throughput):.1f}  {np.std(D2_throughput):.1f}")
print(f"D3 - Accuracy: {np.mean(D3_acc):.3f}  {np.std(D3_acc):.3f}, Throughput: {np.mean(D3_throughput):.1f}  {np.std(D3_throughput):.1f}")
```
