```python
import matplotlib.pyplot as plt
import numpy as np

throughputs = [28.23, 29.08, 30.33, 30.66, 31.77, 31.49, 31.31, 30.45, 29.55, 29.65, 45.81]
accs =        [94.1,   93.6,  93.5,  93.1,  92.8,  92.7,  92.6,  92.6,  92.2,  91.8,  83.3]

# Create x-axis values (experiment numbers)
x = np.arange(len(throughputs))

# Create figure and primary axis
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot throughput on primary y-axis (left)
color1 = '#1f77b4'
ax1.set_xlabel('Number of MucNets', fontsize=12)
ax1.set_ylabel('Throughput (fps)', color=color1, fontsize=12)
line1 = ax1.plot(x, throughputs, color=color1, marker='o', linewidth=2, markersize=6, label='Throughput')
ax1.tick_params(axis='y', labelcolor=color1)
ax1.grid(True, alpha=0.3)

# Create secondary y-axis (right) for accuracy
ax2 = ax1.twinx()
color2 = '#ff7f0e'
ax2.set_ylabel('Accuracy (%)', color=color2, fontsize=12)
line2 = ax2.plot(x, accs, color=color2, marker='s', linewidth=2, markersize=6, label='Accuracy')
ax2.tick_params(axis='y', labelcolor=color2)

# Set x-axis labels
ax1.set_xticks(x)
ax1.set_xticklabels([f'{i+1}' for i in range(len(throughputs))])

# Add title
plt.title('Throughput vs Accuracy for DI1', fontweight='bold', fontsize=14, pad=20)

# Add value annotations
for i, (throughput, acc) in enumerate(zip(throughputs, accs)):
    # Annotate throughput values
    ax1.annotate(f'{throughput:.1f}', (i, throughput), 
                textcoords="offset points", xytext=(0,10), ha='center', 
                fontsize=8, color=color1)
    # Annotate accuracy values
    ax2.annotate(f'{acc:.1f}%', (i, acc), 
                textcoords="offset points", xytext=(0,-15), ha='center', 
                fontsize=8, color=color2)

# Create combined legend
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left', bbox_to_anchor=(0.02, 0.98))

# Adjust layout and show
plt.tight_layout()
plt.show()

# Print some statistics
print("Performance Analysis:")
print("=" * 40)
print(f"Throughput - Min: {min(throughputs):.1f} fps, Max: {max(throughputs):.1f} fps, Mean: {np.mean(throughputs):.1f} fps")
print(f"Accuracy   - Min: {min(accs):.1f}%, Max: {max(accs):.1f}%, Mean: {np.mean(accs):.1f}%")
print(f"Correlation coefficient: {np.corrcoef(throughputs, accs)[0,1]:.3f}")
```
